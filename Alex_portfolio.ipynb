```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
sns.set(color_codes=True)

```

```python
path = r'C:\Users\someone\Desktop\WORK\Data analysis\movies.csv'
movies = pd.read_csv(path)
movies.head()
```

```python
#look for missing data
for col in movies.columns:
    pct_missing = np.mean(movies[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
for col in movies.columns:
    pct_missing = np.sum(movies[col].isnull())
    print('{} - {}'.format(col, pct_missing))
```

```python
movies.dtypes
```

```python
movies['budget'] = movies['budget'].fillna(0).astype('Int64')
movies['gross'] = movies['gross'].fillna(0).astype('Int64')
```

```python
movies
```

```python
#creating columns
movies['New column'] = None
```

```python
movies['New column'] = movies['released'].astype(str).str[:4]
```

```python
#Scatter plot
plt.scatter(x=movies['budget'],y=movies['gross'])
plt.title('Budget Vs Gross earnings')
plt.ylabel('Gross earnings')
plt.xlabel('Budget for film')
plt.show
```

```python
movies1 = movies.loc[(movies["budget"] != 0) & (movies["gross"] != 0)]

sns.regplot(x='budget', y='gross', data=movies1, fit_reg=True, scatter_kws={'color':'blue'}, line_kws={'color':'red'})
plt.xlabel('Budget')
plt.ylabel('Gross')
plt.title('Regression Plot: Gross vs. Budget')
plt.show()
```

```python
#Pearson Kendall and spearman
movies2 =movies[['year','score','votes','budget','gross','runtime']]
movies2.corr() 

```

```python
movies2 =movies[['year','score','votes','budget','gross','runtime']]
movies_corr = movies2.corr() 
sns.heatmap(movies_corr, annot=True)
plt.xlabel('Movies_features')
plt.ylabel('Movies_features')
plt.title('Movies correlation matrix ')
plt.show()
```

```python
movies_numerized = movies

for col_name in movies_numerized.columns:
    if(movies_numerized[col_name].dtype == 'object'):
        movies_numerized[col_name] = movies_numerized[col_name].astype('category')
        movies_numerized[col_name] = movies_numerized[col_name].cat.codes
        
movies_numerized
```

```python
movies_cor = movies_numerized.corr() 
plt.figure(figsize=(20, 8))
sns.heatmap(movies_cor, annot=True)
plt.xlabel('Movies_features')
plt.ylabel('Movies_features')
plt.title('Movies correlation matrix ')
plt.show()
```

```python
movies_cor.corr()
```

```python
correlation_mat = movies_cor 
movies_pairs = correlation_mat.unstack()
movies_pairs
```

```python
corr_sorted = movies_pairs.sort_values()
corr_sorted
```

```python
high_corr = corr_sorted[(corr_sorted) > 0,5]
high_corr
```
